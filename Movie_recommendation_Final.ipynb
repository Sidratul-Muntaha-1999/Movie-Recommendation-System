```python

import pandas as pd
import matplotlib.pyplot as plt

# Movie data
movie_data = {
    'Movie id': list(range(1, 21)),
    'Title': [
        'Inception', 'The Dark Knight', 'Titanic', 'The Matrix', 'The Shawshank Redemption',
        'Pulp Fiction', 'Forrest Gump', 'Gladiator', 'Interstellar', 'The Godfather',
        'Padma Nadir Majhi', 'Chokher Bali', 'Monpura', 'Pather Panchali', 'Shatru',
        'Autograph', 'Shonar Pahar', 'Bhooter Bhabishyat', 'Aparajito', 'Guptodhon Sandhane'
    ],
    'Genre': [
        'Sci-Fi, Thriller', 'Action, Crime', 'Drama, Romance', 'Sci-Fi, Action', 'Drama',
        'Crime, Drama', 'Drama, Romance', 'Action, Drama', 'Sci-Fi, Drama', 'Crime, Drama',
        'Drama, Romance', 'Drama, Romance', 'Drama, Romance', 'Drama', 'Action, Thriller',
        'Drama, Musical', 'Drama, Family', 'Comedy, Fantasy', 'Drama', 'Adventure, Mystery'
    ],
    'Release Year': [
        2010, 2008, 1997, 1999, 1994, 1994, 1994, 2000, 2014, 1972,
        1993, 2003, 2009, 1955, 2011, 2010, 2018, 2012, 1956, 2018
    ],
    'Rating': [
        8.8, 9.0, 7.8, 8.7, 9.3, 8.9, 8.8, 8.5, 8.6, 9.2,
        8.5, 7.8, 8.6, 8.7, 6.9, 8.1, 8.0, 8.2, 8.4, 7.2
    ],
    'Number of Votes': [
        2100000, 2300000, 1800000, 1600000, 2300000,
        1800000, 1700000, 1900000, 2500000, 3000000,
        1200000, 1500000, 5000000, 2800000, 1000000,
        3500000, 6000000, 2500000, 1200000, 1800000
    ],
    'Duration': [
        148, 152, 195, 136, 142, 154, 142, 155, 169, 175,
        126, 167, 135, 125, 150, 137, 111, 123, 110, 133
    ],
    'Director': [
        'Christopher Nolan', 'Christopher Nolan', 'James Cameron', 'The Wachowskis', 'Frank Darabont',
        'Quentin Tarantino', 'Robert Zemeckis', 'Ridley Scott', 'Christopher Nolan', 'Francis Ford Coppola',
        'Goutam Ghose', 'Rituparno Ghosh', 'Giasuddin Selim', 'Satyajit Ray', 'Raj Chakraborty',
        'Srijit Mukherji', 'Parambrata Chatterjee', 'Anik Dutta', 'Satyajit Ray', 'Dhrubo Banerjee'
    ],
    'Description': [
        'A thief who steals corporate secrets through the use of dream-sharing technology.',
        'A vigilante crime fighter faces off against a criminal mastermind known as the Joker.',
        'A seventeen-year-old aristocrat falls in love with a kind but poor artist.',
        'A computer hacker learns from mysterious rebels about the true nature of his reality.',
        'Two imprisoned men bond over a number of years, finding solace and eventual redemption through acts of common decency.',
        'The lives of two mob hitmen, a boxer, a gangster’s wife, and a diner waitress intertwine in four tales of violence and redemption.',
        'The presidencies of Kennedy and Johnson, the Vietnam War, the civil rights movement, and other historical events unfold from the perspective of an Alabama man with an extraordinary life.',
        'A former Roman General sets out to exact vengeance against the corrupt emperor who murdered his family and sent him into slavery.',
        'A team of explorers travel through a wormhole in space in an attempt to ensure humanity\'s survival.',
        'The aging patriarch of an organized crime dynasty transfers control of his clandestine empire to his reluctant son.',
        'The lives of fishermen by the Padma River and explores themes of love, survival, and socio-economic struggles.',
        'A widow’s struggle with love, desire, and social norms in early 20th-century Bengal.',
        'The tragic love between a caretaker and a fisherman’s daughter.',
        'The struggles of a poor family in rural Bengal through the eyes of a young boy.',
        'A righteous police officer stands against corruption and violence while facing personal challenges in a small village.',
        'The lives of a superstar and a struggling director.',
        'The unlikely bond between a lonely old woman and a seven-year-old boy.',
        'Ghosts from different eras living together in a mansion and their efforts to save it from being demolished.',
        'Apu’s journey from childhood to adolescence and his mother’s sacrifices.',
        'A thrilling treasure hunt led by a history professor and his team as they unravel ancient mysteries in Bengal.'
    ],
    'User Ratings': [[] for _ in range(20)],
    'Reviews': [[] for _ in range(20)]
}



# Create a DataFrame
df = pd.DataFrame(movie_data)

# Save the DataFrame to a CSV file
csv_file_path = 'movies_data.csv'
df.to_csv(csv_file_path, index=False)

print(f"CSV file '{csv_file_path}' created successfully.")
print(movie_data)




   #3####################################################################################################################
def add_rating_review():
    movie_id = int(input("Enter Movie ID (1-20): "))
    rating = float(input("Enter your rating (0-10): "))
    review = input("Enter your review: ")

    # Ensure the movie ID exists
    if movie_id >= 1 and movie_id <= 20:
        # Append rating and review to the lists
        df.at[movie_id - 1, 'User Ratings'].append(rating)
        df.at[movie_id - 1, 'Reviews'].append(review)

        # Calculate new average rating
        avg_rating = sum(df.at[movie_id - 1, 'User Ratings']) / len(df.at[movie_id - 1, 'User Ratings'])
        df.at[movie_id - 1, 'Rating'] = avg_rating

        print(f"Rating and review added for {df.at[movie_id - 1, 'Title']}.")
    else:
        print("Invalid Movie ID. Please enter a number between 1 and 20.")

#4 Function to display updated movie data (after rating/review)
def recommend_movies_data():
    print(df[['Movie id', 'Title', 'Rating', 'User Ratings', 'Reviews']])

# Main menu
def main_menu():
    while True:
        print("\n--- Movie Recommendation System ---")
        print("1. Add Ratings and Reviews")
        print("2. Display Movie Data")
        print("0. Exit")

        choice = input("Enter your choice: ").strip()

        if choice == '1':
            add_rating_review()
        elif choice == '2':
            recommend_movies_data()
        elif choice == '0':
            print("Exiting the system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")
recommend_movies_data()

#1 Function to filter movies by Genre, Release Year, or Director
def filter_movies():
    print("\nFilter Movies:")
    print("1. By Genre")
    print("2. By Release Year")
    print("3. By Director")
    filter_choice = input("Enter your choice: ").strip()

    if filter_choice == '1':
        genre = input("Enter genre to filter by (e.g., Sci-Fi): ").strip()
        filtered = df[df['Genre'].str.contains(genre, case=False)]
    elif filter_choice == '2':
        try:
            year = int(input("Enter release year to filter by: ").strip())
            filtered = df[df['Release Year'] == year]
        except ValueError:
            print("Invalid input for year.")
            return
    elif filter_choice == '3':
        director = input("Enter director's name to filter by: ").strip()
        filtered = df[df['Director'].str.contains(director, case=False)]
    else:
        print("Invalid choice.")
        return

    if not filtered.empty:
        print(filtered[['Movie id', 'Title', 'Genre', 'Release Year', 'Rating', 'Director']])
    else:
        print("No movies found with the given filter.")

#2 Function to search movies by title, keywords in descriptions, or genre
def search_movies():
    print("\nSearch Movies:")
    keyword = input("Enter a keyword to search (title, genre, or description): ").strip().lower()
    result = df[df['Title'].str.contains(keyword, case=False) |
        df['Genre'].str.contains(keyword, case=False) |
        df['Description'].str.contains(keyword, case=False)]

    if not result.empty:
        print(result[['Movie id', 'Title', 'Genre', 'Release Year', 'Rating', 'Director']])
    else:
        print("No movies found for the given keyword.")


#5 Function to generate a top 10 list based on Genre, Release Year, or Rating
def generate_top_10():
    print("\nGenerate Top 10 Movies:")
    print("1. By Genre")
    print("2. By Release Year")
    print("3. By Rating")
    choice = input("Enter your choice: ").strip()

    if choice == '1':
        genre = input("Enter genre to get top 10 (e.g., Sci-Fi): ").strip()
        top_movies = df[df['Genre'].str.contains(genre, case=False)].nlargest(10, 'Rating')
    elif choice == '2':
        try:
            year = int(input("Enter release year to get top 10: ").strip())
            top_movies = df[df['Release Year'] == year].nlargest(10, 'Rating')
        except ValueError:
            print("Invalid input for year.")
            return
    elif choice == '3':
        top_movies = df.nlargest(10, 'Rating')
    else:
        print("Invalid choice.")
        return

    print(top_movies[['Movie id', 'Title', 'Genre', 'Release Year', 'Rating', 'Director']])

#6 Function to display graphs (Rating Distribution, Movie Count per Genre)
def display_graphs():
    print("\nDisplaying Graphs:")
    print("1. Rating Distribution")
    print("2. Movie Count per Genre")
    graph_choice = input("Enter your choice: ").strip()

    if graph_choice == '1':
        df['Rating'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
        plt.title('Rating Distribution')
        plt.xlabel('Rating')
        plt.ylabel('Frequency')
        plt.show()
    elif graph_choice == '2':
        genre_counts = df['Genre'].str.split(',').explode().str.strip().value_counts()
        genre_counts.plot(kind='bar', color='skyblue')
        plt.title('Movie Count per Genre')
        plt.xlabel('Genre')
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.show()
    else:
        print("Invalid choice.")

# Main menu function for interaction
def main_menu():
    while True:
        print("\n--- Movie Recommendation System ---")
        print("1. View Movies by Filter")
        print("2. Search Movies")
        print("3. Add Ratings and Reviews")
        print("4. Recommended Movies Based on User Rating History")
        print("5. Generate Top 10 Movies")
        print("6. Display Graphs")
        print("0. Exit")

        choice = input("Enter your choice: ").strip()

        if choice == '1':
            filter_movies()
        elif choice == '2':
            search_movies()
        elif choice == '3':
            add_rating_review()
        elif choice == '4':
            recommend_movies_data()
        elif choice == '5':
            generate_top_10()
        elif choice == '6':
            display_graphs()

        elif choice == '0':
            print("Exiting the system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main_menu()

```
